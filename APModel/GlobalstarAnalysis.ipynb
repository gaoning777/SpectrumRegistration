GlobalStar Analysis
====================

The following analysis examines the time-varying impact of the operation of multiple WiFi 802.11 access points on the performance of specific GlobalStar satellites. You can specify a specific date to use for the analysis as well as a specific satellite in the post-2007 GlobalStar constellation for detailed explication. The analysis at the end of the notebook computes the model over all satellites on the specified date.

The program simulates the motion of the satelite(s) across the 24 hours of the specific date that you specifiy. The RF interference caused by the activity of the WiFi access points, modified to reflect usage patterns across the day, satellite antenna patterns, WiFi access point antenna patterns and clutter models is summed to derive the total RF power due to interference at the satellite. That RF power is then used to compute the "noise rise" at the satellite receiver as well as a change in system capacity to duplex communication on via the satellites.

## Decorate ZCTA Information With Additional Data

We need to approximate the distribution of access points across the United States. We can estimate urban, suburban and rural areas using population density. We'll then assign the number of access points based on the number of access points per area based on the urban/suburban/rural designation similar to the distribution used in the Globalstar / Roberson analysis, which extrapolated from a Google outdoor WiFi deployment in Mountain View, CA.

```python
TotalAPs = 0
TotalPop = 0
NumRural, NumSubUrban, NumUrban = 0, 0 , 0
AreasRural, AreasSubUrban, AreasUrban = 0, 0, 0
SqMiRural, SqMiSubUrban, SqMiUrban = 0, 0, 0
ZeroAps = 0

for zipName in ZCTA.keys():
    z = ZCTA[zipName]
    if z['pop'] <= 1:
        #
        # An extremely rural area, with no APs
        #
        z['aps'] = 0
        NumRural = NumRural + 0
        AreasRural = AreasRural + 1
        SqMiRural = SqMiRural + z['aland_sqmi']
    else:
        #
        # Compute population density
        #
        pd = z['pop'] /  z["aland_sqmi"]                                                      
        #                                                                                                 
        # The type of area is determined by the population density.
        # Although it's not a definitive source, http://greatdata.com/rural-urban-data
        # states the Department of Defense had established the following designations for a ZIP Code:
        # Urban: 3,000+ persons per square mile
        # Suburban: 1,000 ‐ 3,000 persons per square mile
        # Rural: less than 1,000 persons per square mile
        #
        # As that website indicates, this is a course and sometimes inaccurate
        # classification. We'll use cut-offs that make more areas be classified
        # as "urban" and thus increase the number of WiFi access points and is
        # more inline with Census designation of "urban" (>1000 ppl/mi)
        #
        # The value 0.386 is the conversion factor between square miles 
        # and square kilometers. I.e. 1 sq.mi is 1/0.386 or 2.59 sq. km
        #
        if pd < 750:                                                                                      
            z['aps'] = int(z['aland_sqmi'] * 0.05 / 0.386) # 1 AP every 20 sq. mi
            NumRural = NumRural + z['aps']
            AreasRural = AreasRural + 1
            SqMiRural = SqMiRural + z['aland_sqmi']
        elif pd < 1000:                                                                                   
            z['aps'] = int(z['aland_sqmi'] * 10 / 0.386)    # 10 AP's every square km
            NumSubUrban = NumSubUrban + z['aps']
            AreasSubUrban = AreasSubUrban + 1
            SqMiSubUrban = SqMiSubUrban + z['aland_sqmi']
        else:                                                                                             
            z['aps'] = int(z['aland_sqmi'] * 16 / 0.386)    # 16 AP's every square km
            NumUrban = NumUrban + z['aps']
            AreasUrban = AreasUrban + 1
            SqMiUrban = SqMiUrban + z['aland_sqmi']

        TotalAPs = TotalAPs + z['aps']
        TotalPop = TotalPop + z['pop']
    if z['aps'] == 0:
        ZeroAps = ZeroAps + 1
        
print "Total population is", TotalPop, "with", TotalAPs, "AP's"
print "Of the ", len(ZCTA.keys()), "zip codes", \
    ZeroAps, "do not have an AP (", int(1000* ZeroAps / len(ZCTA.keys())) / 10., "%)"
print AreasRural, "rural areas with", NumRural, "AP's over", \
    SqMiRural, "SqMi or ", SqMiRural/ 0.386, "SqKm"
print AreasSubUrban, "suburban areas with", NumSubUrban, "AP's over", \
    SqMiSubUrban, "SqMi or", SqMiSubUrban / 0.386, "SqKm"
print AreasUrban, "urban areas with", NumUrban, "AP's over", \
    SqMiUrban, "SqMi or ", SqMiUrban / 0.386, "SqKm"
```

We describe the antenna pattern using a list of angle offsets and gain specified in dBi. You can specify arbitrary antenna patterns this way. This antenna has a strong null at 0 degrees elevation (i.e. directly up) -- this is done to spread signal horizontially (labeled 270 & 90 degrees) as well as down towards the users. That strong null also has the benefit of reducing interference with the satellite when it is closest to the AP (directly overhead).

```python
# We only need to do the gain up to 90 degrees because of 
# the way that the bearing to the satellite is calculated.
# We include the measurement for the 120-degree angle to simplify
# the logic in gainByAngle

AntennaGainTable = [ (0, 5), (10, 3), (20, 0), 
               (30, -5,), (60, -12), (90, -25), 
               (120,-12) ]

def gainByAngle(angle):
    assert angle >= 0 and angle <= 90
    i = 0
    while angle > AntennaGainTable[i][0]:
        i = i + 1
    return AntennaGainTable[i][1]

print "Here's a sample antenna gain at 10, 25 and 31 degree elevation:", \
   gainByAngle(10), ",", gainByAngle(25), ",", gainByAngle(31)
```

## Describing WiFi duty cycle

WiFi access points do not transmit 100% of the time -- this is impossible due to the WiFi MAC protocol but also because WiFi hotspot activity is driven by human actions. Analysis of the 802.11b/g WiFi MAC shows the duty-cycle can be no more than ~70%.

Extensive measurement studies have shown that WiFi usage depends on the time of the day, which is based on human activity. We assume a peak 40% duty cycle in our model -- this means that over the course of *e.g.* an hour, the WiFi signal would be transmitting 40% of the time for **all** of the access points. In addition, empirical observation of duty cycles suggests that real-world simultaneous use is much lower than maximum theoretical levels enabled by technology standards. One study [link here](http://www.ncbi.nlm.nih.gov/pubmed/23085070) noted that 10% duty cycle is at the higher end of what might be observed. We will use a peak duty cycle of 40% to be conservative and avoid under-estimating the potential effect of simultaneous use.

To allow flexibility in modeling the usage pattern, we specify the duty cycle to depend on the time-of-day at a specific location. The pattern is specified by DutyCycleByHour.

```python
DutyCycleByHour = [
                   (0, 0.05), (6, 0.05), (8, 0.10),
                   (9, 0.20),  (11, 0.40),
                   (12, 0.30), (13, 0.30), (17, 0.40),
                   (18, 0.30), (19, 0.20), (20, 0.30),
                   (21, 0.20), (22, 0.10), (23, 0.05),
                   (24, 0.05), (25, 0.05) # 25th hour used to simplify logic
                   ]
def dutyCycleByHour(hour):
    assert hour >= 0 and hour <= 24      
    i = 0
    while hour > DutyCycleByHour[i][0]:
        i = i + 1
    return DutyCycleByHour[i][1]
```

## Computing the Radio Link

```python
def toMw(dBm):                                                                                          
    return pow(10., (dBm/10.)  )

def toLinear(dB):
    return pow(10., (dB/10))
                                                                                                          
def todBm(mw):                                                                                              
    return 10. * log10(mw) 

def todB(x):                                                                                              
    return 10. * log10(x) 
                                                                                                          
def freqToWave(freq):                                                                                     
    return 3. * 10**8 / freq
```

```python
def pathLoss(dist, mhz = 5125):                                                              
    """Includes frequency specific path loss                                                              
    -- note this is a positive number to be subtracted from                                               
    -- other signals, not a negative number.
    -- dist is in meters and frequency is in Mhz
    """                                                                                                                                                           
    return todBm( ((4 * math.pi * dist) / freqToWave(mhz * 1000. * 1000.))**2 )
```

### Computation of Capacity Reduction Based On Noise Rise

The following computes the change in capacity for a single satellite link based on an increase in the noise rise.
These equations are modeled after the equations in the Globalstar / Roberson analysis. We solve the equation for the combined uplink and down link using an additional noise rise.

Several constants are taken directly from the Roberson report. In particular, that report states that the combined C/N (or Eb/No+Io) with no additional interference from AP's is 1.01, which occurs when the downlink is 209.6dBW/Hz. We recompute the "1.01" using other given terms because the lack of precision in the reported values leads to very small imprecisions in the reported capacity. By computing these from the original values, we get the same results as Roberson's report once rounded.

We use those constants (plus the Eb and No for the downlink) to determine what an "adjusted Lo" would be given the additional WiFi links. This is then used to compute the capacity reduction as done in their computation. That reduction is returned as a percentage (0..100)


```python
processingGain = (1228800. / 2400. )
pgdB = todB( processingGain )

#
# Calculate downlink Eb(no+lo), which is 1.06 in Roberson report
# We get 1.06164538404
#
loDensity = -148.7 - 10*log10(1228800)
print "loDensity", loDensity
#
# This includes some constants from Globalstar
# -168 and -203.9 that we treat as "given"
downEb =  -168 - todB( 1228800 )
print "Original downEb is", downEb
origDownEbNoLo = todB( toLinear(downEb) / (toLinear(-203.9) + toLinear(loDensity)) ) + pgdB
print "Original downlink Eb/(no+lo) is", origDownEbNoLo

#
# Now, calculate overall Eb(No+Lo) given original 19.9 uplink Eb/(no+lo)
# This was 1.01 in Roberson report, we get 1.00526516109
#
origEbNoLo = todB( 1 / ( 1 / toLinear(19.9) + 1 / toLinear(origDownEbNoLo) ) )
print "origEbNoLo", origEbNoLo

pristineEbNoLo = 1 / toLinear(origEbNoLo - pgdB)

eb_dl = toLinear(-168 - 10*log10(1228800))
no_dl = toLinear(-203.9)  # from Globalstar Filing

def capacityChange( nr ):
    """
    Given overall noise rise at the uplink, compute the capacity reduction
    """
    adjustedPristine = (pristineEbNoLo - 1/toLinear(19.9 - nr - pgdB ))    
    reduction = 1 - toLinear( todB(adjustedPristine * (eb_dl ) - no_dl) - loDensity )
    return reduction * 100

#
# Lets print out the capacity change used in the Roberson report
# to verify the math is the same. This should match line AH of Table 2.
#
print "Capacity reduction given a 10dB noise increase is", capacityChange(10)
#
# We also show that the capacity change with 0dB of additional noise
# is zero. If we just use the constants from the Roberson report for
# the target Eb/(no+lo) for the downlink (1.01) there's a small capacity
# reduction of 0.96 at 0 noise rise because of numerical imprecision;
# by recalaculating those values, our model has a 0.0% change for 0dB noise rise,
# indicating agreement with the Robeson report.
#
print "Capacity reduction given a  0dB noise increase is", capacityChange(0)
```

### Function to compute power, noise rise & capacity over US

This function does the core computation for the model. For the specified satellite, the function computes values across time from 0 UTC to 24 UTC in 5 minute increments (12 steps/hour).

During each time period, we compute the added power arising from AP's in each ZCTA zipcode. We use the local time of the location to select the duty cycle of the AP's at this location for this time. As in the earlier examples, we use the ephemeris to determine the distance between the zip code observation point and the satellite nadir. This distance is used to de-rate the satellite antenna from a peak of 3.1dB to 0dB at the edge and then more rapid roll-off beyond that. We use the angle of the satellite above the horizon to determine the AP antenna gain and the clutter values. 

Computing the total power over all satellites is fairly time-consuming.

```python
def computePowerAndRise(date, satName, numOfChannels = 9, daysToAnalyze=1):
    dayStart = ephem.Date(date)
    sat = TLE[satName]
    # Sample 12 times per hour or every 5 minutes of each day
    samplesPerHour = 12.0
    sampleOffsets = range(int(daysToAnalyze * 24 * samplesPerHour))

    PowerAtSat = []
    NoiseRise = []
    ApsActive = []
    CapacityReduction = []
    totalComp = 0
    nullAngle = 0
    nullAps = 0
    prints = 0

    for offset in sampleOffsets:
        utcHourOffset = (offset / samplesPerHour ) % 24  # Adjust to hour in day
        utc = ephem.Date(dayStart + (ephem.hour * offset)/samplesPerHour)
        #
        # Sum the power for this time set in power
        #
        power = 0.0
        totalAps = 0
        #
        # Iterate over all the ZCTA's
        #
        for zipName in ZCTA:
            totalComp = totalComp + 1
            zcta = ZCTA[zipName]
            me = zcta['observer']
            me.date = utc
            sat.compute(me)
            angle = math.degrees( sat.alt )
            dist = sat.range                            # distance in meters
            aps = zcta['aps'] / numOfChannels           # AP's distributed over bands
 

            if angle > AnalyzeHorizonAngle and aps > 0:
                #
                # Compute duty cycle
                #
                localTime = utcHourOffset + zcta['tz']
                if localTime < 0:                       # adjust to local timezone
                    localTime = localTime + 24
                dutyCycle = dutyCycleByHour( localTime )
                
                satLat = math.degrees( float(sat.sublat))
                satLon = math.degrees( float(sat.sublong))
                myLat = float(zcta['lat'])
                myLon = float(zcta['lon'])
                distToNadir = 6373 * distance_on_unit_sphere( satLat, satLon, myLat, myLon )
                #
                # Sat Receiver is 3.1dB at nadir, falls to -3dB at 5800/2 km
                # Without specific information on pattern we approximate the
                # antenna roll-off as 1dB every 725km, and then a rapid roll-off
                # beyond 2900km due to the horn antenna.
                #
                if distToNadir < 725:
                    dBr = 3.1
                elif distToNadir < 1450:
                    dBr = 2.0
                elif distToNadir < 2175:
                    dBr = 1.0
                elif distToNadir <= 2900:
                    dBr = 0.0
                else:
                    dBr = -10 * (distToNadir-2900)/500.
                
                pw = AnalyzeWiFiMw
                pw = pw / 2.0                           # Loss for circular vs. linear polarization
                dBm = todBm( pw * dutyCycle )
                dBi = gainByAngle( angle )
                dBclutter = clutterLossByAngle(angle)
                pl = pathLoss( dist )
                power = power + toMw( dBm + dBi - dBclutter - pl + dBr) * aps
                totalAps = totalAps + aps

        if power == 0:
            PowerAtSat.append( -200.0 )  # no actual power, but to make graphs nicer
            NoiseRise.append( 0 )
            CapacityReduction.append( 0 )
        else:
            PowerAtSat.append( todBm(power) )
            nr = todBm( toMw(OriginalSystemNoisedBm) + power ) - OriginalSystemNoisedBm 
            NoiseRise.append( nr )
            CapacityReduction.append( capacityChange(nr) )
        ApsActive.append( totalAps )
    return sampleOffsets, PowerAtSat, NoiseRise, ApsActive, CapacityReduction
```

```python
# define functions to compute the mean of non-zero elements
# of a list, used to compute average capacity change for
# non-zero changes (i.e. only for affected satellites)
def flatten(l):
    return [item for sublist in l for item in sublist]
def removeZeros(l):
    return filter(lambda a: a != 0, l)
def nonZeroMean(l):
    return np.mean(removeZeros( flatten( l ) ) )
```

### Sample computation of Power, Noise Rise & Capacity For Single Satellite

This computation is similar to the one in Roberson, but using the time-varying WiFi usage and the distances to the AP's. In this example, we divided the total access points across the 12 WiFi channels spanning the 2.4GHz, the UNII-3 and proposed UNII-1 channels.

```python
hours, PowerAtSat, NoiseRise, ApsActive, CapacityChange \
    = computePowerAndRise(AnalyzeDate, AnalyzeSat, numOfChannels=12, daysToAnalyze=AnalyzeDays)
print "Computed ", len(PowerAtSat), "samples across the day"
```

We see almost all the access points operating in this single 20MHz channel. In this example, we expect to see ~300,000 APs at the satellite (~3.2M/12). In certain passes over the US, the satellite see many fewer access points (or doesn't see them for very long) because the satellite is off one or the other coast.

```python
fig = plt.figure()
fig.set_size_inches(12,3)
plt.ylabel("Number of Access Points %s " % (AnalyzeSat))
plt.xlabel("Time (UTC) on %s" % (AnalyzeDate))
plt.axis([0, 24 * AnalyzeDays, 0, 3000000])
highlightDays(AnalyzeDays)
plt.plot( [ (t * 24. * AnalyzeDays)/len(hours) for t in hours], ApsActive)
plt.show()
```

However, although we see almost all the access points operating in this 20MHz channel, many of them do not contribute significant power to the satellite because of the antenna and the long distance between AP's and the satellite.

```python
fig = plt.figure()
fig.set_size_inches(12,3)
plt.ylabel("dBm power at %s " % (AnalyzeSat))
plt.xlabel("Hours since midnight UTC on %s" % (AnalyzeDate))
plt.axis([0, 24 * AnalyzeDays, -160, -100])
highlightDays(AnalyzeDays)
plt.plot([ (t * 24. * AnalyzeDays)/len(hours) for t in hours], PowerAtSat)
plt.show()
```

The resulting noise rise at the satellite receiver is small, bursty and influenced by the time of day both because of the usage pattern of WiFi and the position of the satellite over the United States.

```python
fig = plt.figure()
fig.set_size_inches(12,3)
plt.ylabel("dB Noise Rise at %s " % (AnalyzeSat))
plt.xlabel("Hours since midnight UTC on %s" % (AnalyzeDate))
plt.axis([0,24 * AnalyzeDays,0, 2])
highlightDays(AnalyzeDays)
plt.plot([ (t * 24. * AnalyzeDays)/len(hours) for t in hours], NoiseRise)
plt.text(5, 1.5, \
         "Peak noise rise is %3.2fdB" % max(NoiseRise),\
         fontsize=28)
plt.text(2, 1, \
         "Avg. (non-zero) noise rise is %3.2fdB" % np.mean(removeZeros(NoiseRise)),\
         fontsize=28)
plt.show()
```

We report the peak noise rise for a single satellite across the measurement period, as well as the average non-zero noise rise across the period. We report only non-zero values to reflect the proportion of time that the satellite is in view of the United States.

The noise rise can be used to calculate the satellite capacity reduction.

```python
fig = plt.figure()
fig.set_size_inches(12,3)
plt.ylabel("%% Capacity Reduction at %s " % (AnalyzeSat))
plt.xlabel("Hours since midnight UTC on %s" % (AnalyzeDate))
plt.axis([0,24 * AnalyzeDays,0,20])
plt.plot([ (t * 24. * AnalyzeDays)/len(hours) for t in hours], CapacityChange)
highlightDays(AnalyzeDays)
plt.text(5, 10, \
         "Peak capacity reduction is %3.2f%%" % max(CapacityChange),\
         fontsize=28)
plt.text(1, 5, \
         "Avg. (non-zero) capacity reduction is %3.2f%%" % np.mean(removeZeros(CapacityChange)),\
         fontsize=28)
plt.show()
```

We report the peak capacity reduction for a single satellite across the measurement period, as well as the average capacity reduction across the period. We report only non-zero values to reflect the proportion of time that the satellite is in view of the United States.

This analysis computes the increased noise level assuming that all access points are distributed across those channels. We perform the computation for ALL satellites and overlay the power & noise information on a single graph.
